# p-value

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
import math
```

```python
def gaussian_distribution_predict(x, data):
    """
    Calculate result of x
    
    :param x: value that will calculate gaussian distribution result
    :param data: all values
    
    return  estimated value
    """
    
    mean = np.mean(data)
    std = np.std(data)
    e = math.e
    pi = math.pi
    
    first_part = 1/(std*np.sqrt(2*pi))
    second_part = e**((-(x - mean)**2)/(2*std**2))
    prediction = first_part * second_part
    
    return prediction
```

```python
df = pd.read_csv('./CompanyABCProfit.csv')
data = df['Profit(Rs \'000)'].values
data
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_data = sc.fit_transform(data.reshape(-1,1))
```

```python
estimated_datas = []
for x in scaled_data:
    estimated_datas.append(gaussian_distribution_predict(x, scaled_data))
```

```python
np.shape(estimated_datas)
```

```python
plt.scatter(scaled_data, estimated_datas);
plt.axvline(x = np.mean(scaled_data), linestyle = "--");
```
